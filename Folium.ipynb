Lib for viewing map points

```
import folium
```

**Render Map**

```
map = folium.Map(location=[-5.79448,-35.211])#latitude and longitude
map
```

**Different map styles**

parameter tiles, options: Stamen Toner, Stamen Terrain

```
folium.Map(location=[-5.79448,-35.211],
           tiles = "Stamen Toner"
           )
```

**Zoom Levels**

parameter zoom_start

```
folium.Map(location=[-5.79448,-35.211],
           zoom_start = 12
           )
```

**Add markers**

```
folium.Marker(location = [-5.8434568,-35.1994619],
	            popup = "<i>ECT-UFRN</i>",
	            tooltip = "Click here",
              icon = folium.Icon(color = "black")
	            ).add_to(map)
```

```
map
```

**Circle Marker**

```
folium.CircleMarker(location = [-5.8434568,-35.1994619],
                    radius = 50,
                    color = "green",
                    fill = True,	
                    fill_color = "green"	
     	             ).add_to(map)
```

```
map
```

Add to map location with a click

```
map.add_child(folium.LatLngPopup())
```

**Plot sales map**

```
import pandas as pd
from folium import plugins
```

```
df = pd.read_csv("olist_geolocation_dataset.csv")
df.head()
```

```
df.info()
```

```
df["geolocation_state"].unique()
```

```
df["geolocation_state"].value_counts()
```

```
df_rn = df[df["geolocation_state"] == "RN"]
df_rn
```

```
coordenate = []
for lat,lng in zip(df_rn["geolocation_lat"].values[:5000],df_rn["geolocation_lng"].values[:5000]):
	coordenate.append([lat,lng])
```

**Heat Map**

```
map.add_child(plugins.HeatMap(coordenate))
map
```

**Save map**

```
map.save("map-sales.html")
```
